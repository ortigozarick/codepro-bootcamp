```python
#Bucles infinitos un bucle infinito sucede cuando la condicion siempre es cierta 
#y por lo tanto el bucle no termina de ejecutarse. "Break" es una palabra reservada 
# que nos permite salir de un bucle
numero = 30

while True:
    numero_a_ingresar = input("Ingresa un numero")
    if numero_a_ingresar == numero:
        print("El numero es correcto")
        break
    else:
        print("El numero es incorrecto, intentalo de nuevo")
```

```python
ingrese_contrasenha= int(input("Ingrese su contrasenha"))

while True:
    contrasenha = int(input("Ingresa la contrasenha actual"))
    if ingrese_contrasenha == contrasenha:
        print("La contrasenha es correcta")
        break
    else:
        print("La contrasenha es incorrecta, intentelo de nuevo")
```

```python
#Orden desde 0, 1, 2, 3 y sucesivamente
lista_frutas=['naranja','pera', 'manzana']
print(lista_frutas)
#Lista de boleanos
lista_bool= [True, False]
print(lista_bool)
lista_numeros=[1, 2.5, 3, 4, 5.5]
print(lista_numeros)
lista_cosas= ['zapatos', True, 6.5, 'hola']
print(lista_cosas)

print(lista_frutas[0], lista_cosas[0], lista_bool[1])

#Metodo append es anexar argumentos a las variables
lista_frutas.append("kiwi")
print(lista_frutas)

#Modificar elementos de una lista seleccionando su posicion con [] y =
lista_frutas[1]= "pinhas"

#Metodo para sacar elementos de una lista .remove
lista_frutas.remove('manzana')
print(lista_frutas)

#Metodo para sacar elementos de una lista por posicion .remove
lista_frutas.pop(1)
print(lista_frutas)

# Cantidad de elementos de una lista

print(len(lista_frutas))

cantidad_elementos= len(lista_frutas)
print(cantidad_elementos)
```
