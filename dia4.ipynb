```python
class AnimalDeGranja:
    #atributos de clase
    tipo = "animal de granja"

    #atributos de instancia
    def __init__(self, nombre, color, sonido):
        self.nombre = nombre
        self.color = color
        self.sonido = sonido

    #metodos (acciones de la clase)
    def hablar(self):
        print(f"Hablo y digo {self.sonido}")

    def presentarse(self):
        print(f"Hola soy {self.nombre}")

class Gallina(AnimalDeGranja):
    huevos_por_dia = 1
    
    def poner_huevos(self):
        print(f"Estoy poniendo {self.huevos_por_dia} huevos por dia")

    class Vaca(Gallina):
        patas = 4

        def mujir(self):
            print("Muuuuu")
            
gallina1 = Gallina("Hortencia", "Marron", "Corococo")
gallina1.hablar()
gallina1.presentarse()

vaca = Vaca("Lola", "Blanco y negro", "Muuuu")
vaca.hablar()
vaca.mujir()
vaca.patas
```
